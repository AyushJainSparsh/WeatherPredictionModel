```python
dataset = pd.read_csv(r'forecast_data.csv')
dataset
```

```python
dataset['state'].unique()
```

```python
filter = dataset['state'] == 'Uttar Pradesh'
d1 = dataset.where(filter)
d1.dropna(inplace = True, ignore_index = True)
```

```python
d1
```

```python
drop = ['time_epoch','time','temp_f','is_day' , 'wind_mph', 'wind_degree','wind_dir' ,'pressure_in','precip_mm','precip_in' ,
       'cloud','feelslike_f','windchill_f','heatindex_f','dewpoint_c','dewpoint_f']
d1.drop(drop ,axis = 'columns' , inplace = True)
```

```python
d1.to_csv('Uttar Pradesh.csv' , index = False)
```
